```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# white canvas
canvas = np.ones((400, 400, 3), dtype=np.uint8) * 255

# ellipses (OpenCV logo colors: Blue, Green, Red)
cv2.ellipse(canvas, (200, 120), (60, 60), 0, 0, 360, (255, 0, 0), -1)   # Blue ellipse (top)
cv2.ellipse(canvas, (130, 240), (60, 60), 0, 0, 360, (0, 255, 0), -1)   # Green ellipse (bottom left)
cv2.ellipse(canvas, (270, 240), (60, 60), 0, 0, 360, (0, 0, 255), -1)   # Red ellipse (bottom right)

# "OpenCV" text
cv2.putText(canvas, "OpenCV", (120, 350), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 0, 0), 4, cv2.LINE_AA)

# Display using matplotlib
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('OpenCV Logo Recreation')
plt.show()

#save the image
cv2.imwrite('opencv_logo.png', canvas)
```
