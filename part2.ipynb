```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load images
img1 = cv2.imread('img.png')
img2 = cv2.imread('window2.png')

# dimensions of img1
height, width = img1.shape[:2]

# Resize img2 to match img1
img2_resized = cv2.resize(img2, (width, height)) 

alpha = 0.6  

blend = ((1 - alpha) * img1 + alpha * img2_resized).astype(np.uint8)

# Display using matplotlib
plt.imshow(cv2.cvtColor(blend, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Manual Image Blending')
plt.show()

# save the image
cv2.imwrite('manual_blend.jpg', blend)
```
